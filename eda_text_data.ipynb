```python
categ_features = ['EducationLevel', 'Smoking', 'FamilyHistoryAlzheimers', 'CardiovascularDisease', 'Diabetes', 'Depression', 'HeadInjury', 'Hypertension', 'MemoryComplaints', 'BehavioralProblems', 'Confusion', 'Disorientation', 'PersonalityChanges', 'DifficultyCompletingTasks', 'Forgetfulness']
numeric_features = ['Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity', 'DietQuality', 'SleepQuality', 'SystolicBP', 'DiastolicBP', 'CholesterolTotal', 'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides', 'MMSE', 'FunctionalAssessment', 'ADL']
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(df[numeric_features].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('ÐšÐ¾Ñ€ÐµÐ»ÑÑ†Ñ–Ð¹Ð½Ð° Ð¼Ð°Ñ‚Ñ€Ð¸Ñ†Ñ')
plt.show()
```

```python
def plot_feature_vs_diagnosis(df, col, numeric_features, categ_features):
    plt.figsize=(16, 8)
    if col in numeric_features:
        sns.boxplot(data=df, x="Diagnosis", y=col, hue='Gender')
        plt.title(f"{col} vs Diagnosis")
    elif col in categ_features:
        sns.countplot(data=df, x=col, hue="Diagnosis")
        plt.title(f"{col} distribution by Diagnosis")

    plt.show()

features = [
    "MMSE", "FunctionalAssessment", "ADL",         
    "FamilyHistoryAlzheimers", "Hypertension", 
    "CardiovascularDisease", "Diabetes", "Depression", "Smoking",
    "BMI", "PhysicalActivity", "SleepQuality", "DietQuality"       
]


for col in features:
    plot_feature_vs_diagnosis(df, col, numeric_features, categ_features)
```
